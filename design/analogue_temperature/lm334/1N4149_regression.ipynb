```python
import matplotlib.pyplot as plt
import scipy as sp
import pandas as pd

# Read input file
data = pd.read_csv("diode_test_1n4149.data", delimiter=" ", header=None)

# Update column name for convenience 
data = data.rename(columns={
    0 : "temp_C",
    1 : "Vd_1N4149"
})

lreg = sp.stats.linregress(data["temp_C"], data["Vd_1N4149"])

print(f"Temperature coefficient: {lreg.slope*1e3:.3f} mV/C")

plt.plot(data["temp_C"], lreg.slope*data["temp_C"] + lreg.intercept, 'k--', label=f"{lreg.slope*1e3:.3f} mV/C + {lreg.intercept:.3f} V")
plt.plot(data["temp_C"], data["Vd_1N4149"], '.', label='Sim. Data')
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Diode Forward Voltage (Volts)")
plt.legend()

plt.savefig("diode_test_In4149.png")
```
